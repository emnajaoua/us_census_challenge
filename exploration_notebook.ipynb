# Data Analysis and Exploration

```python
# Education Distribution
df_train['AHGA'].hist()
```

```python
# Age Distribution
sns.distplot(df_train.AAGE)
plt.axvline(df_train.AAGE.mean(), label='mean')
plt.legend()
plt.show()
```

```python
# Instance Weight Distribution
labels = df_train['MARSUPWT'].astype('category').cat.categories.tolist()
replace_map_comp = {'MARSUPWT' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}

print(replace_map_comp)
df_replace = df_train.copy()
df_replace.replace(replace_map_comp, inplace=True)

sns.distplot(df_replace.MARSUPWT, label=replace_map_comp)
plt.axvline(df_replace.MARSUPWT.mean(), label='mean')
plt.legend()
plt.show()
```

```python
# Distribution of Marital status in function of age
labels = df_train['ARACE'].astype('category').cat.categories.tolist()
replace_map_comp = {'ARACE' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
print(replace_map_comp)
df_replace = df_train.copy()
df_replace.replace(replace_map_comp, inplace=True)

plt.pie(df_replace['ARACE'])

plt.axis('equal')
plt.show()
```

```python
# Distribution of Marital status in function of age
df_train.groupby(['AMARITL','salary']).size().unstack().plot(kind='bar',stacked=True, figsize=(20,5))
plt.show()
```
